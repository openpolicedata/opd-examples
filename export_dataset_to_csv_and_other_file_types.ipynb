# Export Dataset to CSV, [Feather](https://arrow.apache.org/docs/python/feather.html), and [Parquet](https://parquet.apache.org/) Files

This notebook shows an example of how to export a dataset to csv, [feather](https://arrow.apache.org/docs/python/feather.html), and [parquet](https://parquet.apache.org/) files. 
It assumes you found the dataset using techniques shown in `finding_datasets.ipynb` and loaded the dataset using `loading_datasets.ipynb`

```python
import openpolicedata as opd
```

```python
# To access the data, create a source using a Source Name (usually a police department name). There is an optional state input to clarify ambiguities.
# We will use the above cell's information for Maryland to choose the agency "Montgomery County" which we select for the source_name

src = opd.Source(source_name="Montgomery County", state="Maryland")
src.datasets.head()
```

```python
# Load traffic stop data for 2021
t = src.load(table_type='TRAFFIC STOPS', year=2021)
```

```python
# Show the first 5 rows of the table
t.table.head(n=5)
# Now you are ready for analyzing the data in the table t.
```

```python
import os
cwd = os.getcwd()
csv_filepath = cwd

print(f"The CSV file will be written to {csv_filepath}. Make sure this path is okay before running the next cell. If the path is not okay then modify csv_filepath.")
```

Files in OPD can be imported and exported to CSV, [feather](https://arrow.apache.org/docs/python/feather.html), and [parquet](https://parquet.apache.org/) data formats. CSV import and export will be demonstrated here but feather and parquet functions also exist with the same inputs.

Here is how to export data with `to_csv`. `to_feather` and `to_parquet` functions are also available.

```python
# Save to CSV. To specify a custom filename, set the filename input
csv_written_filename=t.to_csv(output_dir=csv_filepath)
print(f"The CSV file was written to {csv_written_filename}.")
```

Import the data that was just exported with `load_csv`. `load_feather` and `load_parquet` functions are also available.

```python
# To load data back in from CSV, create a new source and use load_csv
# load_csv usage is similar to load except for the output_dir
# input
src = opd.Source(source_name="Montgomery County", state="Maryland")
t = src.load_csv(year=2021, table_type='TRAFFIC STOPS', output_dir=csv_filepath)
t.table.head()
```
